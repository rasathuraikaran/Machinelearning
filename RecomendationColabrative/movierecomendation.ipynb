```python
import pandas as pd
from scipy import sparse
from sklearn.metrics.pairwise import cosine_similarity
ratings=pd.read_csv("toy_dataset.csv",index_col=0)
```

```python
ratings
```

```python
ratings.fillna(0, inplace=True)
ratings
```

```python
def standardize(row):
    new_row = (row - row.mean())/(row.max()-row.min())
    return new_row

df_std = ratings.apply(standardize).T
print(df_std)
```

```python
item_similarity=cosine_similarity(df_std)
item_similarity
```

```python
item_similarity_df=pd.DataFrame(item_similarity,index=ratings.columns,columns=ratings.columns)
item_similarity_df
```

```python
# lets make prediction
def get_similar_movies(movie_name,user_rating):
  similar_score=item_similarity_df[movie_name]*(user_rating-2.5)
  similar_score=similar_score.sort_values(ascending=False)
  return similar_score

```

```python
get_similar_movies("romantic3", 1)
```

```python
action_lover=[("action1",5),("romantic2",1),("romantic3",1)]
similar_moives=pd.DataFrame()

for movie,rating in action_lover:
  similar_moives=similar_moives.append(get_similar_movies(movie,rating),ignore_index=True)
similar_moives.sum().sort_values(ascending=False)
```
