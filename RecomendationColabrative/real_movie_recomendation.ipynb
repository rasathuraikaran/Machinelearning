```python
import pandas  as pd
```

```python
ratings=pd.read_csv("ratings.csv")
movies=pd.read_csv("movies.csv")
```

```python
ratings=pd.merge(movies,ratings).drop(['genres','timestamp'],axis=1)
ratings
```

```python
user_rating=ratings.pivot_table(index=['userId'],columns=['title'],values='rating')
user_rating
```

```python
user_rating.shape
```

```python
#lets revove the movies which rated less than 10 by users.and fill the remainig with nan with 0
```

```python
user_rating=user_rating.dropna(thresh=10,axis=1).fillna(0)
user_rating
```

```python
item_similarity_df=user_rating.corr(method='pearson')
item_similarity_df.head()
```

```python
# lets make prediction
def get_similar_movies(movie_name,user_rating):
  similar_score=item_similarity_df[movie_name]*(user_rating-2.5)
  similar_score=similar_score.sort_values(ascending=False)
  return similar_score
```

```python
action_lover=[("(500) Days of Summer (2009)",1),("12 Years a Slave (2013)",5)]
similar_moives=pd.DataFrame()

for movie,rating in action_lover:
  similar_moives=similar_moives.append(get_similar_movies(movie,rating),ignore_index=True)
similar_moives.sum().sort_values(ascending=False)
```
